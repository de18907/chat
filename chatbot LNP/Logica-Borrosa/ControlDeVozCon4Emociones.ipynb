```python
import matplotlib
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
# Definición de los rangos de las variables de entrada de los antecedentes (condiciones)
miedo = ctrl.Antecedent(np.arange(0, 11, 1), 'miedo')
rabia = ctrl.Antecedent(np.arange(0, 11, 1), 'rabia')
alegria = ctrl.Antecedent(np.arange(0, 11, 1), 'alegria')
tristeza = ctrl.Antecedent(np.arange(0, 11, 1), 'tristeza')

#Es posible generar las 3 funciones de pertenencia típicas (bajo,medio,alto) para cada variable de entrada usando 
#miVariableDeEntrada.automf(3)
#También se pueden elegir los valores 5 y 7 para obtener una mayor "granularidad" en las funciones de pertenencia

#miedo.automf(3)
#rabia.automf(3)
#alegria.automf(3)
#tristeza.automf(3)

miedo['bajo'] = fuzz.trimf(miedo.universe, [0, 0, 5])
miedo['medio'] = fuzz.trimf(miedo.universe, [0, 5, 10])
miedo['alto'] = fuzz.trimf(miedo.universe, [5, 10, 10])

rabia['baja'] = fuzz.trimf(rabia.universe, [0, 0, 5])
rabia['media'] = fuzz.trimf(rabia.universe, [0, 5, 10])
rabia['alta'] = fuzz.trimf(rabia.universe, [5, 10, 10])

alegria['baja'] = fuzz.trimf(alegria.universe, [0, 0, 5])
alegria['media'] = fuzz.trimf(alegria.universe, [0, 5, 10])
alegria['alta'] = fuzz.trimf(alegria.universe, [5, 10, 10])

tristeza['baja'] = fuzz.trimf(tristeza.universe, [0, 0, 5])
tristeza['media'] = fuzz.trimf(tristeza.universe, [0, 5, 10])
tristeza['alta'] = fuzz.trimf(tristeza.universe, [5, 10, 10])

miedo.view()
rabia.view()
alegria.view()
tristeza.view()
```

```python
volumen = ctrl.Consequent(np.arange(0, 256, 1), 'volumen')
pitch = ctrl.Consequent(np.arange(0, 256, 1), 'pitch')
velocidad = ctrl.Consequent(np.arange(0, 256, 1), 'velocidad')

#Es posible generar las 3 funciones de pertenencia típicas (bajo,medio,alto) para cada variable de entrada usando 
#miVariableDeEntrada.automf(3) 
#También se pueden elegir los valores 5 y 7 para obtener una mayor "granularidad" en las funciones de pertenencia

#volumen.automf(3)
#pitch.automf(3)
#velocidad.automf(3)

volumen['bajo'] = fuzz.trimf(volumen.universe, [0, 0, 123])
volumen['medio'] = fuzz.trimf(volumen.universe, [0, 123, 255])
volumen['alto'] = fuzz.trimf(volumen.universe, [123, 255, 255])

velocidad['baja'] = fuzz.trimf(velocidad.universe, [0, 0, 123])
velocidad['media'] = fuzz.trimf(velocidad.universe, [0, 123, 255])
velocidad['alta'] = fuzz.trimf(velocidad.universe, [123, 255, 255])

pitch['bajo'] = fuzz.trimf(pitch.universe, [0, 0, 123])
pitch['medio'] = fuzz.trimf(pitch.universe, [0, 123, 255])
pitch['alto'] = fuzz.trimf(pitch.universe, [123, 255, 255])

velocidad.view()
```

```python
regla0 = ctrl.Rule(miedo['bajo'] , volumen['medio'])
regla1 = ctrl.Rule(miedo['bajo'] , pitch['medio'])
regla2 = ctrl.Rule(miedo['bajo'] , velocidad['media'])

regla3 = ctrl.Rule(miedo['medio'] , volumen['medio'])
regla4 = ctrl.Rule(miedo['medio'] , pitch['medio'])
regla5 = ctrl.Rule(miedo['medio'] , velocidad['media'])

regla6 = ctrl.Rule(miedo['alto'] , volumen['alto'])
regla7 = ctrl.Rule(miedo['alto'] , pitch['alto'])
regla8 = ctrl.Rule(miedo['alto'] , velocidad['alta'])

#regla1.view()
```

```python
regla10 = ctrl.Rule(rabia['baja'] , volumen['medio'])
regla11 = ctrl.Rule(rabia['baja'] , pitch['medio'])
regla12 = ctrl.Rule(rabia['baja'] , velocidad['media'])

regla13 = ctrl.Rule(rabia['media'] , volumen['medio'])
regla14 = ctrl.Rule(rabia['media'] , pitch['bajo'])
regla15 = ctrl.Rule(rabia['media'] , velocidad['media'])

regla16 = ctrl.Rule(rabia['alta'] , volumen['alto'])
regla17 = ctrl.Rule(rabia['alta'] , pitch['bajo'])
regla18 = ctrl.Rule(rabia['alta'] , velocidad['media'])
```

```python
regla20 = ctrl.Rule(alegria['baja'] , volumen['bajo'])
regla21 = ctrl.Rule(alegria['baja'] , pitch['bajo'])
regla22 = ctrl.Rule(alegria['baja'] , velocidad['baja'])

regla23 = ctrl.Rule(alegria['media'] , volumen['medio'])
regla24 = ctrl.Rule(alegria['media'] , pitch['medio'])
regla25 = ctrl.Rule(alegria['media'] , velocidad['media'])

regla26 = ctrl.Rule(alegria['alta'] , volumen['alto'])
regla27 = ctrl.Rule(alegria['alta'] , pitch['alto'])
regla28 = ctrl.Rule(alegria['alta'] , velocidad['alta'])
```

```python
regla30 = ctrl.Rule(tristeza['baja'] , volumen['alto'])
regla31 = ctrl.Rule(tristeza['baja'] , pitch['alto'])
regla32 = ctrl.Rule(tristeza['baja'] , velocidad['alta'])

regla33 = ctrl.Rule(tristeza['media'] , volumen['medio'])
regla34 = ctrl.Rule(tristeza['media'] , pitch['medio'])
regla35 = ctrl.Rule(tristeza['media'] , velocidad['media'])

regla36 = ctrl.Rule(tristeza['alta'] , volumen['bajo'])
regla37 = ctrl.Rule(tristeza['alta'] , pitch['bajo'])
regla38 = ctrl.Rule(tristeza['alta'] , velocidad['baja'])
```

```python
#Especificamos las reglas que afectan a cada salida
volumen_ctrl = ctrl.ControlSystem([regla0, regla3, regla6, regla10, regla13, regla16,regla20, regla23, regla26, regla30, regla33, regla36])
pitch_ctrl = ctrl.ControlSystem([regla1, regla4, regla7, regla11, regla14, regla17, regla21, regla24, regla27, regla31, regla34, regla37,])
velocidad_ctrl = ctrl.ControlSystem([regla2, regla5, regla8, regla12, regla15, regla18, regla22, regla25, regla28, regla32, regla35, regla38])

#De forma más genérica incluyendo una OR en las condiciones de los antecedentes:
#regla = ctrl.Rule(EntradaA['grande'] | EntradaB['pequeña'], Salida['baja'])
```

```python
#Indicamos al sistema que queremos empezar a usar todo lo anterior
volu = ctrl.ControlSystemSimulation(volumen_ctrl)
pitc = ctrl.ControlSystemSimulation(pitch_ctrl)
velo = ctrl.ControlSystemSimulation(velocidad_ctrl)
```

```python
input_miedo=0.1
input_rabia=10
input_alegria=10
input_tristeza=0.1

#Calculamos el volumen para los valores indicados en input_miedo, input_rabia, input_alegria, input_tristeza
volu.input['miedo'] = input_miedo
volu.input['rabia'] = input_rabia
volu.input['alegria'] = input_alegria
volu.input['tristeza'] = input_tristeza
volu.compute()
print ("Volumen:{0}".format(volu.output['volumen']))
volumen.view(sim=volu)
```

```python
#Calculamos el pitch para los valores indicados en input_miedo, input_rabia, input_alegria, input_tristeza
pitc.input['miedo'] = input_miedo
pitc.input['rabia'] = input_rabia
pitc.input['alegria'] = input_alegria
pitc.input['tristeza'] = input_tristeza
pitc.compute()
print ("Pitch:{0}".format(pitc.output['pitch']))
pitch.view(sim=pitc)
```

```python
#Calculamos la velocidad para los valores indicados en input_miedo, input_rabia, input_alegria, input_tristeza
velo.input['miedo'] = input_miedo
velo.input['rabia'] = input_rabia
velo.input['alegria'] = input_alegria
velo.input['tristeza'] = input_tristeza
velo.compute()
print ("Velocidad:{0}".format(velo.output['velocidad']))
velocidad.view(sim=velo)
```
